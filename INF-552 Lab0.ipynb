### Pandas

#### The read_csv() Method

First, read the documentation about the *read_csv()* method in Pandas (https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.read_csv.html). Then, try to read data from file Salaries.csv to a dataframe, make the column playerID in the csv file as the index column and the first row as the header. Also, skip the second row when reading the file.

````python
# Read Salaries.csv into a DataFrame
df = pd.read_csv("Salaries.csv")

df = pd.read_csv("Salaries.csv", index_col="playerID", header=0, skiprows=[1])
````

#### Indexing and Selecting Data

Select the id of the players who are registered in ATL and HOU and whose salary is higher than one million.

````python
high_salary_players = df[
    ((df["teamID"] == "ATL") | (df["teamID"] == "HOU")) &
    (df["salary"] > 1_000_000)
]

# Show just the IDs (index values)
print(high_salary_players.index.tolist())
````

#### The describe() Method

Calculate the standard Deviation, first quartile, medium, third quartile, mean, maximum, minimum of the salary in team ATL.

````python
atl_salary_stats = df[df["teamID"] == "ATL"]["salary"].describe(
    percentiles=[0.25, 0.5, 0.75]
)

print(atl_salary_stats)
````

#### The iterrows() Method

Create a Python dictionary object whose keys are the headers of the dataframe created in the read_csv() exercise and values are Python list objects that contain data corresponding to the headers. (Here, use the iterrows method to iterate each row of the dataframe and copy it to a dictionary. However, there is a easier way. Learn how the to_dict() method works by yourself later)

````python
# Initialize empty dictionary
data_dict = {col: [] for col in df.columns}

# Fill dictionary using iterrows()
for _, row in df.iterrows():
    for col in df.columns:
        data_dict[col].append(row[col])

print(data_dict)   # optional: see structure
````

#### Create Dataframe Using the Constructor

Read the documentation: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.html#pandas.DataFrame and create a dataframe using pd.DataFrame from the dictionary created in the iterrows() exercise. Change the header to "a", "b", "c", ... at creation time.

````python
new_df = pd.DataFrame(data_dict)

# Number of columns in the new DataFrame
num_cols = new_df.shape[1]

# Generate headers "a", "b", "c", ...
new_headers = [chr(97 + i) for i in range(num_cols)]   # 97 is 'a'

# Rename columns
new_df.columns = new_headers

print(new_df.head())
````

#### ndarray Objects' Attributes

Play with the **ndim, shape, size, dtype, itemsize and data** attribute.

#### The Slice Operation of ndarray Objects

Understand how the slice operation works for 1-D array and 2-D array.

Example:

```python
>>> arr = np.array([[1, 2, 3], [3, 4, 6], [7, 8, 9]])
>>> arr[1:]
array([[3, 4, 6],
       [7, 8, 9]])
>>> arr[1:, 0:2]
array([[3, 4],
       [7, 8]])
```

````python
# 1D slicing
vec = np.array([10, 20, 30, 40, 50])
print("vec[1:4]:", vec[1:4])   # elements 20, 30, 40

# 2D slicing
print("First row:", arr[0, :])     # row 0  [1,2,3]
print("Second column:", arr[:, 1]) # column 1  [2,5,8]
print("Submatrix:\n", arr[0:2, 1:3]) # top-left 22 block

````

#### The Calculation of ndarray Objects

Play with the **argmin(), argmax(), min(), max(), mean(), sum(), std(), dot(), square(), sqrt(), abs(). exp(), sign(), mod()** method.

````python
import numpy as np
import matplotlib.pyplot as plt

# Data
x = np.linspace(-5, 5, num=20)
y1 = np.array([j for j in x])       # y = x
y2 = np.array([j ** 2 for j in x])  # y = x^2
y3 = np.array([j ** 3 for j in x])  # y = x^3

# Create subplots: 1 row, 3 columns
fig, axs = plt.subplots(1, 3, figsize=(15, 4))

# First subplot
axs[0].plot(x, y1, 'r-')
axs[0].set_title("y = x", fontsize=12, fontstyle="italic")
axs[0].set_xlabel("x", fontsize=10)
axs[0].set_ylabel("y", fontsize=10)
axs[0].grid(True)

# Second subplot
axs[1].plot(x, y2, 'b--')
axs[1].set_title("y = x", fontsize=12, fontweight="bold")
axs[1].set_xlabel("x", fontsize=10)
axs[1].set_ylabel("y", fontsize=10)
axs[1].grid(True)

# Third subplot
axs[2].plot(x, y3, 'g:')
axs[2].set_title("y = x", fontsize=12, fontstyle="oblique")
axs[2].set_xlabel("x", fontsize=10)
axs[2].set_ylabel("y", fontsize=10)
axs[2].grid(True)

# Figure title
fig.suptitle("Linear, Quadratic, and Cubic Functions", fontsize=16, fontweight="bold")

plt.tight_layout()
plt.show()
````

(e) Change the limits on x and y axes, **use logarithmic axes to plot**.


copy the code above to the following cell and add code for plotting the parabola.

````python
import numpy as np
import matplotlib.pyplot as plt

# Data
x = np.linspace(0.1, 10, num=50)   # start from 0.1 (avoid log(0) issues)
y = np.array([j ** 2 for j in x])  # parabola y = x^2

# Normal plot with axis limits
plt.plot(x, y, 'bo-')
plt.title("Parabola with Axis Limits")
plt.xlabel("x")
plt.ylabel("y = x^2")
plt.xlim(0, 12)     # x-axis from 0 to 12
plt.ylim(0, 120)    # y-axis from 0 to 120
plt.grid(True)
plt.show()

# Logarithmic y-axis
plt.plot(x, y, 'ro-')
plt.title("Parabola with Logarithmic y-axis")
plt.xlabel("x")
plt.ylabel("y = x^2")
plt.yscale("log")   # log scale on y
plt.grid(True, which="both")  # show grid for major & minor ticks
plt.show()

# Log-log plot (both axes logarithmic)
plt.plot(x, y, 'g^-')
plt.title("Parabola with Log-Log Axes")
plt.xlabel("x (log scale)")
plt.ylabel("y = x^2 (log scale)")
plt.xscale("log")
plt.yscale("log")
plt.grid(True, which="both")
plt.show()
````

#### Seaborn

(a) Append one more numeric feature to the data frame (can be generated randomly), then for team 'ATL', use the seaborn.pairplot to plot scatter plots among all numeric features in the data frame for team. 

````python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Add a random numeric feature (e.g., player rating)
np.random.seed(42)  # for reproducibility
df["rating"] = np.random.randint(50, 100, size=len(df))

# Filter for ATL team
atl = df[df["teamID"] == "ATL"]

# Pairplot among all numeric features for ATL
sns.pairplot(atl, vars=["yearID", "salary", "rating"])
plt.suptitle("ATL Team: Pairplot of yearID, salary, and rating", y=1.02)
plt.show()
````
